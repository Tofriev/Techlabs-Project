```python
import pandas as pd
import os

# Folge der ausgegebenen Anweisung (oben im Outputfenster, vor der ersten Fehlermeldung)
print('Lege den Ordner "data" mit deinen Daten im folgenden Verzeichnis an: ', os.getcwd())
# Erzeuge jetzt einen neuen Ordner mit dem Namen "data" in deinem Arbeitsverzeichnis und hinterlege alle deine Daten darin.

# Im n√§chsten Schritt werden die Daten eingelesen
londoncrime = pd.read_csv("data/london-street.csv")

# Daten bereinigen
lcgeo = londoncrime[["Latitude","Longitude"]]
lcgeo = lcgeo.apply(pd.to_numeric, errors='coerce')
lcgeo = lcgeo.dropna()

#Heatmap generieren
import folium
from folium import plugins
from folium.plugins import HeatMap

#Design der Karte √§ndern mit tiles: 'Stamen Toner' oder 'cartodbdark_matter', oder weglassen
map = folium.Map(location =[51.50939,-0.11832],tiles = 'cartodbdark_matter',
                zoom_start = 13)

HeatMap(lcgeo[:100]).add_to(map)
map
```
